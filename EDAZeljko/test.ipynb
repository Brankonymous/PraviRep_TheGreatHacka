```python
import matplotlib.pyplot as plt

# Sample data for the two scales
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]  # Data for the left y-axis
y2 = [1, 2, 3, 4, 5]      # Data for the right y-axis

# Create a single plot
plt.plot(x, y1, 'b', label='Data 1')  # Blue line for y1
plt.xlabel('X-axis')
plt.ylabel('Data 1', color='b')

# Manually scale the data for the right y-axis
y2_scaled = [val / 10 for val in y2]  # Scale data to match the left y-axis

# Plot the scaled data for the right y-axis
plt.plot(x, y2_scaled, 'r', label='Data 2')  # Red line for y2
plt.ylabel('Data 2 (scaled)', color='r')

# Show a legend for both lines
plt.legend()

plt.title('Dual-Scale Plot')
plt.show()
```
