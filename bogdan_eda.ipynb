```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data/pollen_train.csv')
data.head()
```

```python
cols = data.drop(['Unnamed: 0', 'location', 'date'], axis=1).columns
cols
```

```python
# plot the data with date on x axis and each instance of cols on y axis on same plot

fig, ax = plt.subplots(figsize=(20, 10)) 
for col in cols:
    ax.plot(data['date'], data[col], label=col)
ax.legend()
plt.show()
```

```python
# plot the data with date on x axis and each instance of cols on y axis on same plot using seaborn

fig, ax = plt.subplots(figsize=(20, 10))
sns.lineplot(data=data[cols], ax=ax)
plt.show()
```

```python
alergy_potencial_high = data[['ALNUS', 'AMBROSIA', 'BETULA', 'URTICACEAE']]
alergy_potencial_low = data[['ACER', 'CANNABACEAE', 'CARPINUS', 'FAGUS', 'MORACEAE', 'PINACEAE', 'SALIX', 'TILIA', 'ULMACEAE']]

# create alergy_potencial_mid by dropping alergy_potencial_high and alergy_potencial_low from data
alergy_potencial_mid = data.drop(alergy_potencial_high, axis=1).drop(alergy_potencial_low, axis=1)
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
sns.lineplot(data=alergy_potencial_high, ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
sns.lineplot(data=alergy_potencial_low, ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
sns.lineplot(data=alergy_potencial_mid.drop('Unnamed: 0', axis=1), ax=ax)
plt.show()
```
